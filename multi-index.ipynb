```python
import pandas as pd
import numpy as np
```

```python
df_pob = pd.read_csv('poblacion.csv')
```

```python
df_pob
```

Convertir Year en variable tipo categorica

```python
df_pob['year'] = pd.Categorical(df_pob['year'].apply(str))
```

```python
df_pob.info()
```

```python
#filtro para los paises Aruba y Colombia
idx_filtro = df_pob['Country'].isin(['Aruba','Colombia'])
idx_filtro
```

```python
df_sample = df_pob[idx_filtro]
df_sample
```

```python
df_sample = df_sample.set_index(['Country','year']).sort_index()
df_sample
```

```python
df_sample.loc['Colombia']
```

```python
df_sample.loc['Colombia'].loc['2016',:]
```

La funcion xs permite manejar varios indices

```python
df_sample.xs(['Aruba'])
```

```python
df_sample.xs(['Aruba', '2018'])
```

```python
df_sample.xs('2018', level='year')
```

```python
df_countries = df_pob.set_index(['Country','year']).sort_index()
```
