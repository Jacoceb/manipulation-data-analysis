```python
import pandas as pd
import numpy as np
```

```python
df_left = pd.DataFrame(
    {'X':['x0','x1','x2','x3'],
    'W':['w0','w1','w2','w3'],
    'Y':['y0','y1','y2','y3'],
    'Mix':['y2','y3','a2','a3']},
    index = ['y2','y3','a2','a3'])
```

```python
df_left_dos = pd.DataFrame(
    {'X':['x0','x1','x2','x3'],
    'W':['w0','w1','w2','w3'],
    'Y':['y0','y1','y2','y3']},
    index = ['y2','y3','a2','a3'])
```

```python
df_right = pd.DataFrame(
    {'Z':['z2','z3','z4','z5'],
     'A':['a2','a3','a4','a5'],
     'Y':['y2','y3','y4','y5']},
    index = [2,3,4,5])
```

```python
pd.merge(df_left, df_right)
```

```python
#merge especificando la columna por la que se mezclan
pd.merge(df_left, df_right, how='inner', on='Y')
```

```python
#merge 
pd.merge(df_left, df_right, how='inner', left_on='Mix', right_on='Y')
```

```python
pd.merge(df_left, df_right, how='inner', left_on='Mix', right_on='A')
```

```python
pd.merge(df_left_dos, df_right, how='inner', on='Y')
```

```python
pd.merge(df_left_dos, df_right, how='left', on='Y')
```

```python
pd.merge(df_left_dos, df_right, how='right', on='Y')
```

```python
pd.merge(df_left_dos, df_right, how='outer', on='Y')
```

```python
df_left_tres = pd.DataFrame(
    {'X':['x0','x1','x2','x3'],
    'W':['w0','w1','w2','w3'],
    'Y':['y0','y1','y2','y3'],
    'A':['a0','a1','a2','a3']})
```

```python
pd.merge(df_left_tres, df_right, how='outer', on=['Y','A'])
```

```python
pd.merge(df_left_tres, df_right, how='outer', on='A', suffixes= ['_left','_right'])
```
