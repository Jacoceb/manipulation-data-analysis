```python
# diccionario con dos grupos:
# un grupo con el valor mínimo y valor máximo de la propina
# otro grupo con el valor promedio de las cuentas totales y el promedio en dolares
dict_agg = {'tip':[min, max], 'total_bill':[np.mean, mean_eur2usd]}
dict_agg
```

```python
df.groupby(['sex','time'])[['total_bill','tip']].aggregate(dict_agg)
```

```python
def f_filter(x):
  return mean_eur2usd(x['total_bill'].mean()) > 20
```

```python
# mostrar los datos de las cuentas mayores a 20
df_filtered = df.groupby(['sex','time']).filter(f_filter)
df_filtered
```

# Pivot

```python
# para las variables categoricas de sexo se obtiene el precio promedio de las cuentas en cada hora
df.groupby(['sex','time'])[['total_bill']].mean()
```

```python
df_gp =df.groupby(['sex','time'])[['total_bill']].mean().reset_index()
df_gp
```

```python
df_gp.pivot_table(values='total_bill', index='sex', columns='time')
```

Al hacer el proceso de crear tablas pivote, por defecto, se calcula la media de los datos.

![](https://pandas-docs.github.io/pandas-docs-travis/_images/reshaping_pivot.png)

```python
df.pivot_table(values='total_bill', index='sex', columns='time', aggfunc= np.median) # con la mediana
```

```python
df_pivot = df.pivot_table(values='total_bill', index='sex', columns='time', aggfunc=[np.median, np.std]) # mediana y desviación estándar
df_pivot
```

```python
df_pivot.unstack().reset_index()
```
