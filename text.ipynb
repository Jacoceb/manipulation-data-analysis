```python
import pandas as pd
data = {'names':['Sara Moreno 34',
                 'jUAn GOMez 23',
                 'CArlos mArtinez 89',
                 'Alfredo VelaZques 3',
                 'luis Mora 56',
                 '@freddier #platzi 10',pd.NA]}
```

```python
data
```

```python
df = pd.DataFrame(data)
df
```

```python
#convertir textos en minuscula
df['names'].str.lower()
```

```python
#mayuscula
df['names'].str.upper()
```

```python
#solo la primera letra en mayuscula
df['names'].str.capitalize()
```

```python
#contar longitud de cadenas
df['names'].str.len()
```

```python
#separar las cadenas en un elemento de un arreglo
df['names'].str.split(' ')
```

```python
df['names'].str.split('#')
```

```python
#seleccionar primeros cinco caraceteres de la cadena
df['names'].str[:5]
```

```python
#ultimos cinco
df['names'].str[-5:]
```

```python
#reemplazar la palabra "Alfredo" por "Antonio"
df['names'].str.replace('Alfredo','Antonio')
```

```python
#encontrar secuencia de texto
df['names'].str.findall('ara')
```

```python
#filtro basado en una secuencia de texto
df['names'].str.contains('or')
```

```python
#numero de ocurrencias de un caracter en especifico
df['names'].str.lower().str.count('a')
```

```python
#extraer caracteres numericos
df['names'].str.extract('([0-9]+)', expand=False)
```

```python
#eliminar la mencion con "@"
df['names'].str.replace('@[^\s]+','')
```
